```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['deck', 'embark_town', 'embarked'], axis=1)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# droping all rows with missing data points
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df = df[['pclass', 'sex', 'age', 'survived']]
```

```python
df
```

```python
X = df.drop('survived', axis =1)
Y = df['survived']
```

```python
from sklearn import preprocessing
```

```python
lb = preprocessing.LabelBinarizer()
```

```python
X['sex'] = lb.fit_transform(X['sex'])
# 1:m 0:f
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression()
```

```python
LR_model.fit(X_train, Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier()
```

```python
dtree_model.fit(X_train, Y_train)
```

```python
from sklearn.ensemble import BaggingClassifier
#BC_model = BaggingClassifier()
```

```python
BC_model =BaggingClassifier(base_estimator=None, n_estimators=100, max_samples=1.0, max_features=1.0, bootstrap=True, bootstrap_features=False, oob_score=False, warm_start=False, n_jobs=None, random_state=None, verbose=0)
# default BaggingClassifier(base_estimator=None, n_estimators=10, max_samples=1.0, max_features=1.0, bootstrap=True, bootstrap_features=False, oob_score=False, warm_start=False, n_jobs=None, random_state=None, verbose=0)
```

```python
BC_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
## Accuracy REPORT - LogicticRegression model
print('\nLogistic Regression model\n')
# Accuracy Score
print(accuracy_score(Y_test, LR_model.predict(X_test)))
# Classification Report
print(classification_report(Y_test, LR_model.predict(X_test)))
# Confusion Matrix
print(confusion_matrix(Y_test, LR_model.predict(X_test)))
```

```python
## Accuracy REPORT - DescisionTreeClassifier model
print('\nDecision Tree Classifier\n')
# Accuracy Score
print(accuracy_score(Y_test, dtree_model.predict(X_test)))
# Classification Report
print(classification_report(Y_test, dtree_model.predict(X_test)))
# Confusion Matrix
print(confusion_matrix(Y_test, dtree_model.predict(X_test)))
```

```python
## Accuracy REPORT - BaggingClassifier model
print('\nBagging Classifier - Ensemble learning\n')
# Accuracy Score
print(accuracy_score(Y_test, BC_model.predict(X_test)))
# Classification Report
print(classification_report(Y_test, BC_model.predict(X_test)))
# Confusion Matrix
print(confusion_matrix(Y_test, BC_model.predict(X_test)))
```

```python
from sklearn.model_selection import ParameterGrid

from sklearn.utils import check_random_state
rng = check_random_state(0)

grid = ParameterGrid({"max_samples": [0.5, 1.0],
                          "n_estimators": [100]
                          "max_features": [1, 2, 3],
                          "bootstrap": [True, False],
                          "bootstrap_features": [True, False]})
```

```python
from sklearn.linear_model import Perceptron
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier


for base_estimator in [None,
                           LogisticRegression(),
                           Perceptron(),
                           DecisionTreeClassifier(),
                           KNeighborsClassifier(),
                           SVC(),
                           MLPClassifier()]:
        for params in grid:
            BC_model = BaggingClassifier(base_estimator=base_estimator,
                              #random_state=rng,
                              **params)
            BC_model.fit(X_train, Y_train)
            
            ## Accuracy REPORT - BaggingClassifier model
            print('\n')
            print(base_estimator)
            print('\n')
            # Accuracy Score
            print(accuracy_score(Y_test, BC_model.predict(X_test)))
            # Classification Report
            print(classification_report(Y_test, BC_model.predict(X_test)))
            # Confusion Matrix
            print(confusion_matrix(Y_test, BC_model.predict(X_test)))
```

```python
log_clf = LogisticRegression()
rnd_clf = BaggingClassifier()
svm_clf = SVC()


for clf in (log_clf, rnd_clf, svm_clf):
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_test)
    print(accuracy_score(Y_test, Y_pred))
```
